```python
import os
from google.colab import userdata
userdata.get('GROQ_API_KEY')
from dotenv import load_dotenv
load_dotenv()
from langchain_groq import ChatGroq

model = ChatGroq(api_key=" " ,  model ='llama-3.3-70b-versatile')

```

```python

from langchain_community.tools import DuckDuckGoSearchResults



def search_tool(query:str):
  """ This tool help for web search"""

  search = DuckDuckGoSearchResults()
  return search.invoke(query)

# result = search_tool('what is ai agent')

# print(result)








```

```python
def add_num(a:float , b: float)->float:
  """This tool return the addition of two tools"""

  return a + b



def multiply(a :float , b : float)->float:
  """This tool returns the multiplication of two numbers"""

  return a * b




```

```python
from langgraph.prebuilt import create_react_agent


research_agent = create_react_agent(
    model = model ,
    tools = [search_tool],
    name = "research_agent",
    prompt = "You are a search agent . works on current event . dont answer math related questions"

)
```

```python
from langgraph.prebuilt import create_react_agent


Math_agent = create_react_agent(
    model = model ,
    tools = [add_num , multiply],
    name = "Math_agent",
    prompt = "You are a math agent . only work for math related queries"
)

```

```python
!pip install langgraph_supervisor
```

```python
from langgraph_supervisor import create_supervisor

supervisor = create_supervisor(
    [research_agent , Math_agent] ,
    name = "supervisor" ,
    model = model ,
    prompt=(
        "You are a team supervisor managing a research expert and a math expert. "
        "For current events, use research_agent. "
        "For math problems, use math_agent."
    )
)

graph = supervisor.compile()

result = graph.invoke({
    'messages' : [{
        'role' : 'user',
        'content' : 'what is the weather of mumbai and what will be if i addede 10 in  mumbai weather and multiply by 5'
    }]
})

print(result['messages'][-1].content)
```
